```python
male=pd.read_csv("Deaths-All-Male.csv")
female=pd.read_csv("Deaths-All-Female.csv")
df=pd.read_csv("convert.csv")
def convert_country(compare,dataset):
    #把男生跟女生的資料處理分開
    a=["Congo","Democratic Republic of the Congo","The Gambia","South Korea","North Korea",\
       "Slovakia","England","Yemen"]
    b=["Congo, Rep.","Congo, Dem. Rep.","Gambia","Korea, Dem. Rep.","Korea, Rep.",\
       "Slovak Republic","United Kingdom","Yemen, Rep."]
    for i in range(len(a)):
        compare.Location[compare.Location==a[i]]=b[i]
    compare_list=sorted(set(list(compare.Location)))

    country_name=list(dataset["country"])
    continent=list(dataset["continent"])

    data=compare.sort_values(by=["Location","Year"])
    country,year,value,area=[],[],[],[]
    for k in range(len(country_list)):
        if country_list[k] not in set(list(data.Location)):
            continue
        temp=data[data.Location==country_list[k]]
        country.extend([country_list[k]]*27)
        year.extend(list(range(1990,2017)))
        #print(temp.iloc[0:27]["value"])
        value.extend(temp.iloc[0:27]["Value"])
        area.extend([continent[country_name.index(country_list[k])]]*27)
    df=pd.DataFrame({"country":country,"year":year,"value":value,"continent":area})
    df=df.sort_values(by=["country","year"])
    return df
male=convert_country(male,df)
female=convert_country(female,df)

temp,temp2=list(male.value[:]),list(female.value[:])
df1=pd.DataFrame({"male":temp,"female":temp2})
dataset=pd.concat([df,df1],axis=1)
print(dataset.head(n=4))
```

```python
years = list(range(1990,2017))
# make list of continents
continents = []
for continent in dataset['continent']:
    if continent not in continents:
        continents.append(continent)
# make figure
figure = {
    'data': [],
    'layout': {},
    'frames': []
}

# fill in most of layout
figure['layout']['xaxis'] = {'range': [min(dataset.female), max(dataset.female)], 'title': 'Female'}
figure['layout']['yaxis'] = {'range': [min(dataset.male), max(dataset.male)],'title': 'Male'}
figure['layout']['hovermode'] = 'closest'
figure['layout']['sliders'] = {
    'args': [
        'transition', {
            'duration': 400,
            'easing': 'cubic-in-out'
        }
    ],
    'initialValue': '1900',
    'plotlycommand': 'animate',
    'values': years,
    'visible': True
}
figure['layout']['updatemenus'] = [
    {
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 500, 'redraw': False},
                         'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                'transition': {'duration': 0}}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'direction': 'left',
        'pad': {'r': 10, 't': 87},
        'showactive': False,
        'type': 'buttons',
        'x': 0.1,
        'xanchor': 'right',
        'y': 0,
        'yanchor': 'top'
    }
]

sliders_dict = {
    'active': 0,
    'yanchor': 'top',
    'xanchor': 'left',
    'currentvalue': {
        'font': {'size': 20},
        'prefix': 'Year:',
        'visible': True,
        'xanchor': 'right'
    },
    'transition': {'duration': 300, 'easing': 'cubic-in-out'},
    'pad': {'b': 10, 't': 50},
    'len': 0.9,
    'x': 0.1,
    'y': 0,
    'steps': []
}

# make data
year = 1900
for continent in continents:
    dataset_by_year = dataset[dataset['year'] == year]
    dataset_by_year_and_cont = dataset_by_year[dataset_by_year['continent'] == continent]

    data_dict = {
        'x': list(dataset_by_year_and_cont['female']),
        'y': list(dataset_by_year_and_cont['male']),
        'mode': 'markers',
        'text': list(dataset_by_year_and_cont['country']),
        'marker': {
            'sizemode': 'area',
            'sizeref': 200000,
            #'size': [10000]*len(dataset_by_year_and_cont['country'])
            #list(dataset_by_year_and_cont['pop'])
        },
        'name': continent
    }
    figure['data'].append(data_dict)
    
# make frames
for year in years:
    frame = {'data': [], 'name': str(year)}
    for continent in continents:
        dataset_by_year = dataset[dataset['year'] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[dataset_by_year['continent'] == continent]

        data_dict = {
            'x': list(dataset_by_year_and_cont['female']),
            'y': list(dataset_by_year_and_cont['male']),
            'mode': 'markers',
            'text': list(dataset_by_year_and_cont['country']),
            'marker': {
                'sizemode': 'area',
                'sizeref': 200000,
                #'size': [10000]*len(dataset_by_year_and_cont['country'])
            },
            'name': continent
        }
        frame['data'].append(data_dict)

    figure['frames'].append(frame)
    slider_step = {'args': [
        [year],
        {'frame': {'duration': 300, 'redraw': False},
         'mode': 'immediate',
       'transition': {'duration': 300}}
     ],
     'label': year,
     'method': 'animate'}
    sliders_dict['steps'].append(slider_step)

    
figure['layout']['sliders'] = [sliders_dict]
```
